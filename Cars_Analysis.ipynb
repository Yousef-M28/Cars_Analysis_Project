### Analyzing Cleaning the data

```python
# summary statistics
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Filling Nan Values
df['horsepower'].fillna(df['horsepower'].median , inplace=True)
```

```python
df['name'].unique()
```

```python
# Inserting a new column with the type of car
Car_Type = df["name"].str.split(" ", n = 1, expand = True)[0]
df.insert(1, 'Car_type', Car_Type)
df.head()
```

```python
df['Car_type'].unique()
```

```python
df.Car_type.value_counts()
```

```python
df.name.value_counts()
```

```python
# Editing some wrong car_type names
df.replace({'Car_type':{"maxda":"mazda" , 
                        "toyouta":"toyota" ,
                        "vokswagen":"volkswagen" ,
                        "vw":"volkswagen",
                        "mercedes-benz":"mercedes",
                        "chevroelt":"chevrolet",
                        "chevy":"chevrolet"}},inplace=True)
df['name'].unique()
```

```python
df = df[~df['Car_type'].isin(['hi', 'capri'])]
df
```

```python
df.Car_type.value_counts()
```

```python
car_origin_count = df.groupby(['Car_type','origin'], as_index=False).size().sort_values(by='size', ascending=False)
car_origin_count
```

```python
origin_count = df.groupby(['origin'],as_index=False).size().sort_values(by='origin', ascending=False)
origin_count
```

```python
# Saving the data file
df.to_excel('Cars_dataset.xlsx',sheet_name='Sheet1', index=False)
```
